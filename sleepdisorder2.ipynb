```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Sleep_disorder_data.csv')
```

```python
df
```

```python
df.info()
```

```python
df['Sleep Disorder'].unique()
```

```python
df = df.drop('Person ID',axis=1)
```

```python
df
```

```python
sns.boxplot(df['Age'])
```

```python
df['Blood Pressure'] = df['Blood Pressure'].str.split('/')
```

```python
df
```

```python
df['Blood Pressure'][0]
```

```python
a = []
b = []
for i in df['Blood Pressure']:
    a.append(i[0])
    b.append(i[1])
```

```python
df['Systolic_BP'] = a
```

```python
df['Diastolic_BP'] = b
```

```python
df
```

```python
df = df.drop('Blood Pressure',axis=1)
```

```python
df
```

```python
sd = df.pop('Sleep Disorder')
```

```python
df
```

```python
df['Sleep Disorder'] = sd
```

```python
df
```

```python
df['Sleep Disorder'].unique()
```

```python
df = df.dropna(subset=['Sleep Disorder'])
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
col = ['Gender','Occupation','BMI Category','Sleep Disorder']
for i in col:
    df[i] = le.fit_transform(df[i])
```

```python
df
```

```python
sns.scatterplot(data=df,x=df['Systolic_BP'],y=df['Diastolic_BP'],hue='Sleep Disorder')
plt.title('Systolic_BP vs Diastolic_BP by Sleep Disorder')
plt.show()
```

```python
sns.boxplot(df)
```

# To see if sleep disorders are associated with getting more or less sleep.

```python
sns.boxplot(data=df, x='Sleep Disorder', y='Sleep Duration')
plt.title('Sleep Duration vs Sleep Disorder')
plt.show()
```

# ðŸ“ˆ Insight: 
# Insomniacs tend to sleep less and more consistently short hours.
# Sleep apnea patients have more variable sleep duration â€” some sleep little, others long (possibly compensating), but quality may be low.

```python
sns.boxplot(data=df, x='Sleep Disorder', y='Stress Level')
plt.title('Stress Level vs Sleep Disorder')
plt.show()
```

```python
df.groupby('Sleep Disorder')['Stress Level'].median()
```

# ðŸ“ˆ Insight: 
# Stress is more consistent (less variable) in insomnia, but more scattered and extreme in sleep apnea.
# Both disorders are linked to high stress, but sleep apnea shows a stronger and more unpredictable association.

```python
sns.scatterplot(data=df, x='Sleep Duration', y='Heart Rate', hue='Sleep Disorder')
plt.title('Heart Rate vs Sleep Duration by Sleep Disorder')
plt.show()
```

```python
sns.countplot(data=df, x='Occupation', hue='Sleep Disorder')
plt.title('Sleep Disorders Across Occupations')
plt.xticks(rotation=45)
plt.show()
```

```python
pd.crosstab(df['BMI Category'], df['Sleep Disorder']).plot(kind='bar', stacked=True)
plt.title('Sleep Disorder by BMI Category')
plt.ylabel('Count')
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df, vars=['Sleep Duration', 'Stress Level', 'Heart Rate', 'Daily Steps'], hue='Sleep Disorder')
plt.suptitle('Pair Plot of Features by Sleep Disorder', y=1.02)
plt.show()
```

```python
sns.histplot(data=df, x='Age', hue='Sleep Disorder', kde=True, multiple='stack')
plt.title('Age Distribution by Sleep Disorder')
plt.show()
```
